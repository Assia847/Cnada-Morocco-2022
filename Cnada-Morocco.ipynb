```python
squads_morocco = squads[squads['Team'] == 'Morocco']
del squads_morocco['Team'], squads_morocco['Club'], squads_morocco['League'], squads_morocco['WC Goals']
squads_morocco.set_index('ID', inplace=True)
```

```python
# Create bars
plt.bar(squads_morocco['Position'], squads_morocco['Goals'])

# Create names on the x-axis
plt.xticks(squads_morocco['Position'])
plt.title("Goals per position's player- morroco")
# Show graphic
plt.show()
```

```python
# Create bars
plt.bar(squads_morocco['Position'], squads_morocco['Caps'])

# Create names on the x-axis
plt.xticks(squads_morocco['Position'])
plt.title("Number of caps per position's player - morocco")
# Show graphic
plt.show()
```

```python
world_mor_home= world_cups[world_cups['Away Team']=='Morocco']
world_mor_home
```

```python
world_mor_away= world_cups[world_cups['Home Team']=='Morocco']
world_mor_away
```

```python
win_1=0
draw_1=0
lost_1=0
for i  in range(len(world_mor_home['Home Goals'])):
  if world_mor_home['Home Goals'].iloc[i] > world_mor_home['Away Goals'].iloc[i] :
    win_1=win_1+1
  elif world_mor_home['Home Goals'].iloc[i] == world_mor_home['Away Goals'].iloc[i]:
    draw_1= draw_1+1
  else:
      lost_1= lost_1+1

print("Number of defeats:", lost_1)
print("number of wins:", win_1)
print("number of draws:", draw_1)
```

```python
for i  in range(len(world_mor_away['Home Goals'])):
  if world_mor_away['Home Goals'].iloc[i] < world_mor_away['Away Goals'].iloc[i] :
    lost_1= lost_1+1
  elif world_mor_away['Home Goals'].iloc[i] == world_mor_away['Away Goals'].iloc[i]:
    draw_1= draw_1+1
  else:
      win_1=win_1+1

print("Number of defeats:", lost_1)
print("number of wins:", win_1)
print("number of draws:", draw_1)
```
